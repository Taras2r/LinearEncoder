```python
import csv
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
path = "linEnc.csv"
file = open(path, 'r')
reader = csv.DictReader(file)

period = []
angle = []

for row in reader:
    period.append(float(row['period']))
    angle.append(float(row['angle']) / 100.0)

plt.clf()
plt.close()
plt.figure(dpi = 120)
plt.plot(period, angle, 'r', linestyle='dotted')
plt.grid()
plt.show()
```
