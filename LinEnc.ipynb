```python
import csv
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
path = "linEnc.csv"
file = open(path, 'r')
reader = csv.DictReader(file)

period = []
counts = []
angle_us = []
angle_counts = []
angle_mm = []
rod_mm = []
rod_mmDataSheet = []

for row in reader:
    period.append(float(row['period']) / 100.0)
    counts.append(float(row['counts']))
    angle_us.append(float(row['angle_us']) / 100.0)
    angle_counts.append(float(row['angleCounts']) / 100.0)
    angle_mm.append(float(row['angleMM']) / 100.0)
    rod_mm.append(float(row['MM']) / 100.0)
    rod_mmDataSheet.append(float(row['MMDataSheet']) / 100.0)

plt.clf()
plt.close()
plt.figure(dpi = 120)
plt.plot(counts, angle_mm, 'r', linestyle='--')
plt.xlabel("Start/Stop Period, 48MHz Counter")
plt.ylabel("Pitch Angle, °")
# plt.plot(counts, angle_counts, 'b', linestyle='dotted')
# plt.plot(counts, rod_mm)
plt.grid()
plt.show()

plt.figure(dpi = 120)
plt.plot(period, angle_us, 'r', linestyle='--')
plt.xlabel("Start/Stop Period, μs")
plt.ylabel("Pitch Angle, °")
plt.grid()
plt.show()

plt.figure(dpi = 120)
plt.plot(counts, rod_mm, 'r', linestyle='--', label="Calc. Coef.")
plt.plot(counts, rod_mmDataSheet, 'b', linestyle=':', label="DataSheet Coef.")
plt.xlabel("Start/Stop Period, 48MHz Counter")
plt.ylabel("Magnet Position, mm")
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
